### Gender 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Get the counts for each nationality
counts = df['Gender'].value_counts()

# Calculate the percentages for each nationality
percentages = counts / counts.sum() * 100

# Sort the nationalities by percentage
percentages = percentages.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
# Plot the nationality percentages using Seaborn
sns.set_style("darkgrid")
index = ['Male','Female']
sns.barplot(x=index, y=percentages)
plt.title('Distribution of Nationalities')
plt.xlabel('Gender')
plt.ylabel('Percentage')
plt.show()
```

### Example voice 

```python
import IPython
import librosa
import matplotlib.pyplot as plt
```

```python
IPython.display.Audio("/home/ubuntu/voxceleb/wav/id10002/0_laIeN-Q44/00001.wav")
```

```python
# Load audio file
audio_file, sr = librosa.load('/home/ubuntu/voxceleb/wav/id10002/0_laIeN-Q44/00001.wav')
```

```python
# Plot waveform
plt.figure(figsize=(12,4))
plt.title('Waveform')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.plot(audio_file)
plt.show()
```

By examining this plot, you can see the overall shape of the audio signal, its loudness or volume, and any changes in volume over time. It can also help you identify any irregularities or distortions in the audio, such as clipping or saturation.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Compute spectrogram
spectrogram = librosa.feature.melspectrogram(y=audio_file, sr=sr)

# Plot spectrogram
plt.figure(figsize=(12,4))
librosa.display.specshow(librosa.power_to_db(spectrogram, ref=np.max),
                         y_axis='mel', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.tight_layout()
plt.show()
```

By examining the spectrogram, you can see the frequency content of the audio signal over time. Brighter colors represent higher energy or amplitude of the corresponding frequency components. The spectrogram can help you identify various features of the audio signal, such as pitch, harmonics, and noise.
